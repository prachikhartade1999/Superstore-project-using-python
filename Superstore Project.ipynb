## 2. Bi Variate Analysis

```python
#group by()
```

```python
cat
```

```python
#1. Display Region wise --- i. Total sales    ii.Total profit

#Context: 'Region' , 'Sales'

#1. Total Region wise Sales

df.groupby('Region')['Sales'].sum().round()
```

```python
# 2. % Total Sales

np.round(df.groupby('Region')['Sales'].sum().round(2)   /   df['Sales'].sum().round() * 100 , 2 ) 
```

```python
#3. Total Region wise Profit

df.groupby('Region')['Profit'].sum().round(2)
```

```python
#4. Region wise Profit contribution

np.round(df.groupby('Region')['Profit'].sum() / df['Profit'].sum() * 100 , 2 ) 
```

```python
# 6. Display 'Segment' wise 
#    i. Total Sales 
#    ii Total Profit.
```

```python
# Segment wise 'Sales'

df.groupby('Segment')['Sales'].sum().round(2)
```

```python
# Segment wise 'Profit'

df.groupby('Segment')['Profit'].sum().round(2)
```

```python
# 7.  Display  city wise  sales. 

df.groupby('City')['Sales'].sum().nlargest(5)
```

```python
# 8. Display Regiowise total sales of each Segment.

df.groupby(['Region', 'Segment'])['Sales'].sum().unstack().round()
```
